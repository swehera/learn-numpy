```python
# “1 থেকে 5 এর মধ্যে 6টা number বানাও (equal gap দিয়ে)”
arr = np.linspace(1,5,6) # np.linspace(start, stop, num)   
arr #  gap = (stop - start) / (num - 1) 
```
